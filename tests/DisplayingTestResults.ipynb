# Displaying Test Results

```python
from ndreg import *
%matplotlib inline
```

Suppose we ran LDDMM registration between a input image and reference image and have a resulting displacement field

```python
inImg = imgRead("in.img")
refImg = imgRead("ref.img")
field = imgRead("field.vtk")
```

We can check the results using the imgShowResults command.

```python
imgShowResults(inImg, refImg, field)
```

The first and second rows show the input image and deformed input images respectively.  The 4th row shows the reference image. The 3rd row is a checkerboard composite of the deformed input and reference images. The last row shows the deformed grid.  Notice how the grid is periodic.
